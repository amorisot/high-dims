This plot is surprising for a couple of reasons:

1) It shows that our model thinks that every pair of sentences looks the almost the same, with cosine similarity greater than 0.98. Put another way, it projects all sentences into a remarkably narrow chunk of the high dimensional space in which it operates.

2) It also shows that despite this the model is still capable of some form of semantic understanding: there is a clear correlation between higher machine-labeled cosine similarity and higher human-labeled semantic similarity.

Let's play around with 1) to see if we can figure out why this happens.

# Embeddings that all look the same

All of our sentence embeddings seem to have incredibly high cosine similarities, no matter the original content of the sentence. 

Do we have reason to believe that this is just a property of high dimensional vectors? We can find out by making some! Let's generate 512 embeddings `a` drawn from a 768-dimensional multivariate Gaussian distribution, and find their cosine similarity with 512 embeddings `b` drawn from the same distribution.

```python
a, b = np.random.normal(0, 1, (512, 768)), np.random.normal(0, 1, (512, 768))
```

```python
plt.hist(np.sum(a*b/(norm(a, axis=-1, keepdims=True)*norm(b, axis=-1, keepdims=True)), axis=-1))
plt.show()
```

As we can see, most embeddings don't seem to be too similar to one another, which is what we expected.

But this doesn't reflect how sentence embeddings are made. Sentence embeddings are __sums__ of embeddings. Even if the original embeddings are random and have small cosine similarity scores, doesn't mean their sums do. 

Let's generate the same embeddings `a` and `b` from before, except this time we get the cosine similarties of their respective sums (along the row dimension, so we end up with a 768 dimensional embedding). We do this 500 times and plot the histogram: 

```python
cos_sims = []
for i in range(500):
    a, b = np.random.normal(0, 1, (512, 768)), np.random.normal(0, 1, (512, 768))
    a, b = np.sum(a, axis=0), np.sum(b, axis=0)
    cos_sims.append(np.sum(a*b/(norm(a)*norm(b))))
```

```python
plt.hist(cos_sims)
plt.show()
```

It looks remarkably similar to the previous plot! How odd... 

Well, one thing we could rectify is the assumption that the embeddings are drawn from a perfect random normal. We know from the literature (e.g. this delightful paper: https://arxiv.org/abs/1909.00512) that embeddings get closer together in space as they go through the transformer. What if we incorporated this into our calculations, and instead drew our `a` and `b` embeddings from a normal distribution with non-zero mean?

Let's say, rather than have $\mu$=0, we try using $\mu$=0.3.

```python
mu = 0.3
a, b = np.random.normal(mu, 1, (512, 768)), np.random.normal(mu, 1, (512, 768))
```

```python
plt.hist(np.sum(a*b/(norm(a, axis=-1, keepdims=True)*norm(b, axis=-1, keepdims=True)), axis=-1))
plt.show()
```

Ah, interesting, so the embeddings are now generally closer together. What about their sum? Let's try the same procedure as before, but with our new $\mu$.

```python
cos_sims = []
for i in range(500):
    a, b = np.random.normal(mu, 1, (512, 768)), np.random.normal(mu, 1, (512, 768))
    a, b = np.sum(a, axis=0), np.sum(b, axis=0)
    cos_sims.append(np.sum(a*b/(norm(a)*norm(b))))
```

```python
plt.hist(cos_sims)
plt.show()
```

Woah! If you sum them they become incredibly close together!

I guess there's a way to make sense of this geometrically: if, starting at a lamppost, a drunkard randomly takes a step left/right with 50/50 chance, then even after a day of stumbling, you'd still expect him to not have strayed far from the lamppost. But if he takes a step left with a 51% chance, and a step right with 49% chance, after a day of stumbling you'd expect him to be very far from the lamppost. 

If we push this analogy into higher dimensions, the same principle still applies. If the model has a bias --no matter how slight!-- towards creating token embeddings that point in a particular direction of the 768-dimensional space, then the sum of those embeddings will likely point in the direction of that bias (this is probably the right place to name-drop the central limit theorem).

# A possible fix

This should be a straightforward fix then: let's try normalising our embeddings. 

However, let's go hard, and say that the initial embeddings (prior to the sum) need to have very high cosine similarity. This will more accurately reflect real-life token embeddings. So in addition to setting our normal distribution's $\mu$=0.3, we also set $\sigma$=0.1!

```python
mu, sigma = 0.3, 0.1
a, b = np.random.normal(mu, sigma, (512, 768)), np.random.normal(mu, sigma, (512, 768))
a -= np.mean(a, axis=0)
b -= np.mean(b, axis=0)
plt.hist(np.sum(a*b/(norm(a, axis=-1, keepdims=True)*norm(b, axis=-1, keepdims=True)), axis=-1))
plt.show()
```

Looking good! Despite making the variance much smaller, we still end up with low cosine similarities! Let's try the sum now. 

```python
cos_sims = []
for i in range(500):
    a, b = np.random.normal(mu, sigma, (512, 768)), np.random.normal(mu, sigma, (512, 768))
    a -= np.mean(a, axis=0)
    b -= np.mean(b, axis=0)
    a, b = np.sum(a, axis=0), np.sum(b, axis=0)
    cos_sims.append(np.sum(a*b/(norm(a)*norm(b))))
```

```python
plt.hist(cos_sims)
plt.show()
```

Also looking good! Now finally, the most important part, does this normalisation preserve true similarities between embeddings? Let's replace the first embeddings in `a` and `b` with extremely similar embeddings (with cosine similarity =1), and see how the normalisation effects them. We'll try a couple of ways, one where the embeddings are equal to $\mu$, and one where they are $\epsilon$ away (here we choose $\epsilon$=0.1).

```python
epsilon = 0.1

plt.subplot(1, 2, 1)
a, b = np.random.normal(mu, sigma, (512, 768)), np.random.normal(mu, sigma, (512, 768))
a[0] = (mu + epsilon)*np.ones(a[0].shape)
b[0] = (mu + epsilon)*np.ones(b[0].shape)
a -= np.mean(a, axis=0)
b -= np.mean(b, axis=0)
cos_sims = np.sum(a*b/(norm(a, axis=-1, keepdims=True)*norm(b, axis=-1, keepdims=True)), axis=-1)
plt.hist(cos_sims, bins=20)
plt.title("With $\epsilon$")

plt.subplot(1, 2, 2)
a, b = np.random.normal(mu, sigma, (512, 768)), np.random.normal(mu, sigma, (512, 768))
a[0] = mu*np.ones(a[0].shape)
b[0] = mu*np.ones(b[0].shape)
a -= np.mean(a, axis=0)
b -= np.mean(b, axis=0)
cos_sims = np.sum(a*b/(norm(a, axis=-1, keepdims=True)*norm(b, axis=-1, keepdims=True)), axis=-1)
plt.hist(cos_sims, bins=20)
plt.title("Without $\epsilon$")


plt.show()
```

Weird! And kind of interesting.

I have a hypothesis as to why this would be the case, but would love to hear your opinions. 

But I think the conclusion is that it's worth trying to normalise token embeddings before summing them up and turning them into sentence embeddings.

Some code for getting the token embeddings of a Huggingface DistilBERT model:

```python
import torch
from transformers import DistilBertModel, DistilBertTokenizer

model_class, tokenizer_class, pretrained_weights = (DistilBertModel, DistilBertTokenizer, 'distilbert-base-uncased')

# Load pretrained model/tokenizer
tokenizer = tokenizer_class.from_pretrained(pretrained_weights)
model = model_class.from_pretrained(pretrained_weights)

vocab = model.embeddings.word_embeddings(torch.LongTensor(np.arange(30522))).detach().numpy()
print(vocab.shape)
```

```python
vocab_shuffle = np.copy(vocab)
np.random.shuffle(vocab_shuffle)

cos_sims = np.sum(vocab_shuffle * vocab / (norm(vocab_shuffle, axis=-1, keepdims=True) * norm(vocab, axis=-1, keepdims=True)), axis=-1)

plt.hist(cos_sims)
plt.show()
```
